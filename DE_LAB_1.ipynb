```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
html_content = """
<html>
<head>
    <title>Online Bookstore</title>
</head>
<body>
    <div class="book">
        <h2>Title: The Alchemist</h2>
        <p>Author: Paulo Coelho</p>
        <p>Price: ₹350</p>
    </div>
    <div class="book">
        <h2>Title: Atomic Habits</h2>
        <p>Author: James Clear</p>
        <p>Price: ₹499</p>
    </div>
    <div class="book">
        <h2>Title: Ikigai</h2>
        <p>Author: Héctor García</p>
        <p>Price: ₹299</p>
    </div>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(html_content, 'lxml')
```

```python
books = soup.find_all('div', class_='book')
book_list = []

for book in books:
    title = book.find('h2').text.split(':')[1].strip()
    author = book.find_all('p')[0].text.split(':')[1].strip()
    price = book.find_all('p')[1].text.split(':')[1].strip()
    book_list.append({'Title': title, 'Author': author, 'Price': price})
```

```python
df = pd.DataFrame(book_list)
df
```
